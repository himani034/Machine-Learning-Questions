```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

employee = pd.read_csv("employee_attrition.csv")
sales = pd.read_csv("Sales.csv")
telecom = pd.read_csv("telecom_churn.csv")

print(employee.info())
print(employee.isnull().sum())

plt.figure(figsize=(6,4))
sns.countplot(data=employee, x="Department", hue="Attrition")
plt.title("Employee Attrition by Department")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(data=employee, x="Attrition", y="Monthly_Income")
plt.title("Monthly Income vs Attrition")
plt.show()

print(sales.info())
print(sales.isnull().sum())

sales["Total_Sales"] = sales["Quantity"] * sales["Price"]

product_sales = sales.groupby("Product")["Total_Sales"].sum().sort_values(ascending=False)
print(product_sales)

product_sales.plot(kind="bar", figsize=(7,4), title="Sales by Product")
plt.show()

print(telecom.info())
print(telecom.isnull().sum())

plt.figure(figsize=(5,4))
sns.countplot(data=telecom, x="Churn")
plt.title("Churn Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(data=telecom, x="Contract_Type", hue="Churn")
plt.title("Churn by Contract Type")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(data=telecom, x="Churn", y="Monthly_Charges")
plt.title("Monthly Charges vs Churn")
plt.show()
```
