```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
article_text = """
Artificial Intelligence (AI) is revolutionizing education and research. 
Students often struggle with reading long academic articles. AI-powered research assistants can summarize key points, 
extract insights, answer questions, and provide concise explanations. These tools use Large Language Models (LLMs) 
trained on vast amounts of text to understand content and generate meaningful summaries.
"""
```

```python
def summarize_text(text, max_length=100):
    inputs = tokenizer("summarize: " + text, return_tensors="pt", truncation=True)
    outputs = model.generate(**inputs, max_length=max_length)
    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return summary

summary = summarize_text(article_text)
print("Summary:\n", summary)
```

```python
few_shot_text = """
Example 1:
Article: AI improves healthcare diagnostics.
Summary: AI helps doctors diagnose diseases better.

Example 2:
Article: ML aids in weather forecasting.
Summary: ML improves weather predictions.

Now summarize this article:
""" + article_text

few_shot_summary = summarize_text(few_shot_text)
print("Few-Shot Summary:\n", few_shot_summary)
```

```python
tokens = tokenizer.tokenize(article_text)
print("Tokens:", tokens)
print("Total tokens:", len(tokens))
```

```python
def research_assistant(article):
    prompt = f"""
    Summarize the article professionally:
    - 3 key insights
    - 1 actionable takeaway

    Article: {article}
    """
    return summarize_text(prompt, max_length=150)

output = research_assistant(article_text)
print(output)
```

```python
# Step 0: Install required libraries
!pip install transformers sentencepiece torch --quiet

# Step 1: Import libraries
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Step 2: Load the model (flan-t5-small or larger if needed)
model_name = "google/flan-t5-small"  # for better output use "flan-t5-base" or "flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Step 3: Summarization function
def summarize_text(text, max_length=250):
    """Generates text summary from input text"""
    inputs = tokenizer("summarize: " + text, return_tensors="pt", truncation=True)
    outputs = model.generate(**inputs, max_length=max_length)
    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return summary

# Step 4: Optimized research assistant function
def research_assistant(article):
    """
    Generates:
    - Short summary (2-3 sentences)
    - 3 key insights in bullet points
    - 1 actionable recommendation
    """
    prompt = f"""
    You are an AI research assistant. Read the article carefully and generate a professional response.
    1. Provide a short summary (2-3 sentences)
    2. List 3 key insights in bullet points
    3. Give 1 actionable recommendation

    Article: {article}
    """
    return summarize_text(prompt, max_length=300)  # increased length for better output

# Step 5: Test with example article
article_text2 = """
Machine Learning can help students by providing personalized learning paths. 
AI tutors can analyze performance and suggest areas of improvement. 
Automation of grading can save time for educators.
"""

output = research_assistant(article_text2)
print("Task 5 – AI Research Assistant Output:\n")
print(output)
```
