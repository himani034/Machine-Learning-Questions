```python
# PROMPT ENGINEERING
# Prompt Design Principle -> CLEAR (Context, Length, Examples, Audience, Role)
# Key Prompt technique
# 1. Zero-shot -> Zero-shot prompting means asking a model to complete a task directly using only instructions, without providing sample inputs and outputs.
# 2. Few-shot -> Few-shot prompting is a technique where the model is given a few examples in the prompt to help it understand how to perform the task.
# 3. Chain-of-Thought -> Chain-of-Thought prompting is a technique in prompt engineering where the model explains the intermediate reasoning steps required 
#                        to reach the final answer.It helps the model solve complex problems more accurately.
# 4. Role-based -> Role-based prompting is a technique in prompt engineering where the user assigns a specific role (such as teacher, doctor, or marketing
#                  expert) to the AI so that it generates responses from that perspective.
```

```python
#  Scenario: University Library Search
# Imagine you’re building a smart search system for a university library.
# - Problem: Students often type queries like “fast dog running” or “canine sprinting”, but
# the library catalog has books titled “The dog runs fast”. Traditional keyword search misses these connections.
# - Solution: Use embeddings to capture semantic meaning instead of exact words.

from sentence_transformers import SentenceTransformer

# Choose from popular models:
# "all-MiniLM-L6-v2" → 384 dims, lightweight, fast
# "all-mpnet-base-v2" → 768 dims, stronger semantic understanding
# "BAAI/bge-m3" → 1024 dims, multilingual + dense/sparse retrieval
# "sentence-transformers/e5-large-v2" → 1024 dims, top performer on MTEB

model_name = "all-MiniLM-L6-v2"   # swap this to test other models
model = SentenceTransformer(model_name)

texts = [
    "The dog runs fast",
    "A quick canine sprints",
    "Python is a programming language"
]

embeddings = model.encode(texts)

print(f"Model: {model_name}")
print(f"Embedding shape: {embeddings.shape}")   # e.g. (3, 384) for MiniLM
print(f"First vector sample: {embeddings[0][:10]}")  # show first 10 values
```

```python
# Scenario: Hospital Patient Record Search
# - Problem: Doctors and nurses often type queries like “heart attack” or “chest pain emergency”, but the hospital’s 
# electronic health record (EHR) system stores diagnoses as “acute myocardial infarction”. Traditional keyword search 
# fails because medical staff use everyday language while records use clinical terminology.
# - Solution: Use embeddings to capture semantic meaning. Embeddings understand that “heart attack” and “acute myocardial 
# infarction” are related concepts. This allows the search system to return the right patient records even when the query 
# wording doesn’t match the stored terminology.
    
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

# Hospital patient records (clinical terminology)
records = [
    "Patient diagnosed with acute myocardial infarction",
    "Patient suffering from hypertension",
    "Patient has type 2 diabetes",
    "Patient admitted with pneumonia"
]

# Doctor's search query (everyday language)
query = "heart attack"

# Convert text into embeddings
record_embeddings = model.encode(records)
query_embedding = model.encode([query])

# Compute similarity between query and records
similarity_scores = cosine_similarity(query_embedding, record_embeddings)

# Find the best matching record
best_match_index = similarity_scores.argmax()

print(f"Model: {model_name}")
print(f"Query: {query}")
print(f"Embedding shape: {record_embeddings.shape}")   # e.g. (4, 384)
print(f"Similarity scores: {similarity_scores}")
print(f"Most relevant record: {records[best_match_index]}")
print(f"First vector sample: {embeddings[0][:10]}") 
```

```python
# Scenario: E‑Commerce Product Search
# - Problem: Shoppers often type queries like “cheap running shoes” or “affordable sneakers for jogging”, but the catalog lists products as
# “Budget-fuse marketing terms.
# - Solution: Use embeddings to capture semantic meaning. Embeddings understand that “cheap running shoes” and “budget-friendly athletic footwear” 
# are related concepts. This ensures customers find the right products even when their wording doesn’t match the catalog descriptions.
# from sentence_transformers import SentenceTransformer

# Choose from popular models:
# "all-MiniLM-L6-v2" → 384 dims, lightweight, fast
# "all-mpnet-base-v2" → 768 dims, stronger semantic understanding
# "BAAI/bge-m3" → 1024 dims, multilingual retrieval

model_name = "all-MiniLM-L6-v2"   # you can swap models to test performance
model = SentenceTransformer(model_name)

# Product catalog (marketing terminology)
products = [
    "Budget-friendly athletic footwear",
    "Premium leather formal shoes",
    "Wireless noise-cancelling headphones",
    "Lightweight cotton summer t-shirt"
]

# Customer search query (everyday language)
query = "cheap running shoes"

# Generate embeddings
product_embeddings = model.encode(products)
query_embedding = model.encode([query])

# Compute similarity
from sklearn.metrics.pairwise import cosine_similarity
similarities = cosine_similarity(query_embedding, product_embeddings)

# Find best match
best_match_index = similarities.argmax()

print(f"Model: {model_name}")
print(f"Query: {query}")
print(f"Embedding shape: {product_embeddings.shape}")
print(f"Similarity scores: {similarities}")
print(f"Best matching product: {products[best_match_index]}")
print(f"First vector sample: {embeddings[0][:10]}") 
```
