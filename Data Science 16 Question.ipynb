```python
import pandas as pd
df = pd.read_csv("employee_performance.csv")

print("Original Dataset:")
print(df)
exceptional = df[df["Score"] > 90]
print("\nEmployees with Score > 90:")
print(exceptional)

sales_dept = df[df["Department"] == "Sales"]
print("\nSales Department Records:")
print(sales_dept)

sorted_by_score = df.sort_values(by="Score", ascending=False)
print("\nSorted by Score (Descending):")
print(sorted_by_score)

sorted_dept_score = df.sort_values(by=["Department", "Score"], ascending=[True, False])
print("\nSorted by Department and then Score:")
print(sorted_dept_score)

avg_dept_score = df.groupby("Department")["Score"].mean()
print("\nAverage Score per Department:")
print(avg_dept_score)

max_quarter_score = df.groupby("Quarter")["Score"].max()
print("\nMaximum Score per Quarter:")
print(max_quarter_score)
```
