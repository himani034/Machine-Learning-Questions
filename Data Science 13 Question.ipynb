```python
import pandas as pd
import matplotlib.pyplot as plt
attrition = pd.read_csv("employee_attrition - Sheet1.csv")
print(attrition.shape)
print(attrition.head())
print(attrition.isna().sum())

dept_counts = attrition["Department"].value_counts()
print(dept_counts)
gender_counts = attrition["Gender"].value_counts()
print(gender_counts)

plt.figure(figsize=(8,5))
dept_counts.plot(kind="bar")
plt.title("Attrition Count by Department")
plt.xlabel("Department")
plt.ylabel("Number of Attrition Cases")
plt.show()


import pandas as pd
import matplotlib.pyplot as plt

# Load Sales data
sales = pd.read_csv("Sales1 - Sheet1.csv")

# Inspect
print(sales.shape)
print(sales.head())
print(sales.isna().sum())

# Fill missing data
sales["Price"] = sales["Price"].fillna(sales["Price"].median())

# Total revenue per product
sales["Revenue"] = sales["Quantity"] * sales["Price"]

top_products = sales.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
print(top_products)

# Region performance
top_regions = sales.groupby("Region")["Revenue"].sum()
print(top_regions)

# Plot top products
plt.figure(figsize=(8,5))
top_products.plot(kind="bar")
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.show()



import pandas as pd
import matplotlib.pyplot as plt

# Load churn data
churn = pd.read_csv("telecom_churn - Sheet1.csv")

# Basic info
print("Dataset Shape:", churn.shape)
print(churn.head())
print("Missing Values:\n", churn.isna().sum())

# Churn rate
churn_rate = churn["Churn"].value_counts(normalize=True) * 100
print("\nChurn Rate (%):")
print(churn_rate)

# Churn by contract type
churn_contract = churn.groupby("Contract")["Churn"].value_counts(normalize=True).unstack()
print("\nChurn by Contract Type:")
print(churn_contract)

# Plot churn proportion
plt.figure(figsize=(6,4))
churn["Churn"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Churn Proportion")
plt.ylabel("")
plt.show()
```

```python
# Question 3:
# A college is tracking student attendance across the first four months of the semester: 
# - January: 85 students present on average 
# - February: 90 students 
# - March: 95 students 
# - April: 88 students 
# The academic coordinator wants to visualize the attendance trend to see whether students are attending classes more regularly or # if there are fluctuations. 
# You are asked to: 
# - Plot a line chart showing the monthly attendance trend. 
# - Label the axes and add a title to make the chart clear. 
# - Interpret the chart: 
# - Which month had the highest attendance? 
# - Did attendance steadily increase, or were there ups and downs? 
# - What might explain the dip in April?
import matplotlib.pyplot as plt
months = ["January", "February", "March", "April"]
attendance = [85, 90, 95, 88]
plt.plot(months, attendance, marker='o')
plt.xlabel("Months")
plt.ylabel("Average Student Attendance")
plt.title("Monthly Student Attendance Trend")

plt.show()
# March had the highest attendance
# It increased first, then dropped in April
# The dip in April may be due to exams, fatigue, or reduced student motivation
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("employee_performance.csv")

print("===== ORIGINAL DATA =====")
print(df)

# -------------------------------
# FILTERING
# -------------------------------

# Employees with Score > 90 (exceptional performers)
exceptional = df[df["Score"] > 90]
print("\n===== Employees with Score > 90 =====")
print(exceptional)

# Records for Sales department only
sales_dept = df[df["Department"] == "Sales"]
print("\n===== Sales Department Records =====")
print(sales_dept)

# -------------------------------
# SORTING
# -------------------------------

# Sort by Score (descending)
sorted_by_score = df.sort_values(by="Score", ascending=False)
print("\n===== Sorted by Score (Descending) =====")
print(sorted_by_score)

# Sort by Department first, then Score
sorted_by_dept_score = df.sort_values(by=["Department", "Score"], ascending=[True, False])
print("\n===== Sorted by Department then Score =====")
print(sorted_by_dept_score)

# -------------------------------
# GROUPING
# -------------------------------

# Average score per department
avg_by_department = df.groupby("Department")["Score"].mean()
print("\n===== Average Score per Department =====")
print(avg_by_department)

# Maximum score per quarter
max_by_quarter = df.groupby("Quarter")["Score"].max()
print("\n===== Maximum Score per Quarter =====")
print(max_by_quarter)
```
