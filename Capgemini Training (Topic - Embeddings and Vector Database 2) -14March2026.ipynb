```python
# Simple cosine similarity example
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

a = np.array([0.2,0.8,-0.3,0.5])
b = np.array([0.1,0.9,-0.2,0.4])

print(cosine_similarity(a,b))
```

```python
# Scenario: Healthcare Chatbot – Patient Symptom Matching
# Imagine you’re designing a healthcare chatbot that helps patients find possible conditions based on their symptoms. Both patients and conditions are
# represented as vectors of features (e.g., fever, cough, fatigue, headache).
# - Patient’s symptom vector:
# [0.2,0.8,-0.3,0.5] → mild fever, strong cough, no fatigue, moderate headache.
# - Condition’s symptom profile vector:
# [0.1,0.9,-0.2,0.4] → mild fever, strong cough, little fatigue, moderate headache.
# The chatbot uses cosine similarity to measure how closely the patient’s symptoms align with the condition’s profile

import numpy as np
patient = np.array([0.2, 0.8, -0.3, 0.5])
condition = np.array([0.1, 0.9, -0.2, 0.4])
cos_sim = np.dot(patient, condition) / (np.linalg.norm(patient) * np.linalg.norm(condition))

print("Cosine Similarity:", cos_sim)
```

```python
# #Scenario: E‑Commerce Product Recommendations
# # Imagine you’re building a smart recommendation engine for an online store.
# # - Problem: Customers often search with phrases like “comfortable running shoes”,
#  but the catalog has items
# #  titled “lightweight sneakers for jogging”. Traditional keyword
#  search struggles to connect these.
# # - Solution: Use embeddings to capture semantic meaning of product 
# descriptions and customer queries.

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Customer query embedding (semantic meaning of "comfortable running shoes")
query_embedding = np.array([0.1, 0.8, 0.5])

# Product catalog embeddings (semantic meaning of product descriptions)
product_embeddings = {
    "lightweight sneakers for jogging": np.array([0.2, 0.7, 0.6]),
    "formal leather shoes": np.array([0.9, 0.1, 0.2]),
    "sandals for summer": np.array([0.3, 0.4, 0.5])
}

# Compute similarity between query and each product
for product, embedding in product_embeddings.items():
    similarity = np.dot(query_embedding, embedding)  # fast similarity check
    cosine_sim = cosine_similarity([query_embedding], [embedding])[0][0]  # normalized similarity
    print(f"{product} → Dot Product: {similarity:.3f}, Cosine Similarity: {cosine_sim:.3f}")

# Recommendation: pick the product with highest similarity
best_match = max(product_embeddings.items(), key=lambda x: np.dot(query_embedding, x[1]))
print("\nRecommended Product:", best_match[0])
```

```python
# Scenario: Online Learning Platform Recommendations
# Imagine you’re building a smart recommendation engine for an online education platform.
# - Problem: Students often search with phrases like “beginner-friendly Python tutorials”, but the course catalog has items titled “Introduction to 
# Programming with Python”. Traditional keyword search struggles to connect these because the words don’t match exactly.
# - Solution: Use embeddings to capture the semantic meaning of both student queries and course descriptions. By comparing embeddings with dot product 
# or cosine similarity, the system can recommend the most relevant courses even when the wording is different

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Student query embedding (semantic meaning of "beginner-friendly Python tutorials")
query_embedding = np.array([0.2, 0.8, 0.5])  # example vector

# Course catalog embeddings (precomputed for simplicity)
course_embeddings = {
    "Introduction to Programming with Python": np.array([0.3, 0.7, 0.6]),
    "Advanced Python for Data Science": np.array([0.1, 0.9, 0.4]),
    "Java Programming Basics": np.array([0.8, 0.2, 0.1])
}

# Compute similarity between query and each course
for course, embedding in course_embeddings.items():
    dot_sim = np.dot(query_embedding, embedding)  # fast similarity check
    cos_sim = cosine_similarity([query_embedding], [embedding])[0][0]  # normalized similarity
    print(f"{course} → Dot Product: {dot_sim:.3f}, Cosine Similarity: {cos_sim:.3f}")

# Recommendation: pick the course with highest dot product
best_course = max(course_embeddings.items(), key=lambda x: np.dot(query_embedding, x[1]))
print("\nRecommended Course:", best_course[0])
```

```python
# Scenario: Hospital FAQ Assistant
# A hospital wants to build a chatbot that helps patients quickly find answers to common questions.
# - Problem: Patients ask questions in different ways. For example:
# - “How do I book an appointment?”
# - “What’s the process to schedule a doctor visit?”
# - “Can I see a physician tomorrow?”
# Traditional keyword search might miss these connections.
# - Solution: Convert FAQs and patient queries into embeddings, then measure similarity to find the closest match.
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

query_embedding = np.array([0.3, 0.7, 0.5])
faq_embeddings = {
    "How do I book an appointment?": np.array([0.3, 0.7, 0.6]),
    "What’s the process to schedule a doctor visit?": np.array([0.2, 0.8, 0.5]),
    "Can I see a physician tomorrow?": np.array([0.4, 0.6, 0.4]),
    "What are the hospital visiting hours?": np.array([0.1, 0.3, 0.9])
}
for faq, embedding in faq_embeddings.items():
    dot_sim = np.dot(query_embedding, embedding)
    cos_sim = cosine_similarity([query_embedding], [embedding])[0][0] 
    print(f"{faq} → Dot Product: {dot_sim:.3f}, Cosine Similarity: {cos_sim:.3f}")
best_faq = max(faq_embeddings.items(), key=lambda x: np.dot(query_embedding, x[1]))
print("\nRecommended FAQ:", best_faq[0])
```
