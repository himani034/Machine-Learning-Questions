```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

employee = pd.read_csv("employee_attrition.csv")

print(employee.head())
print(employee.info())
print(employee.isnull().sum())
```

```python
sns.countplot(data=employee, x="Attrition")
plt.title("Overall Employee Attrition")
plt.show()
```

```python
sns.countplot(data=employee, x="Department", hue="Attrition")
plt.title("Attrition by Department")
plt.show()
```

```python
sns.boxplot(data=employee, x="Attrition", y="Monthly_Income")
plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
sns.boxplot(data=employee, x="Attrition", y="Years_at_Company")
plt.title("Years at Company vs Attrition")
plt.show()
```
